```r
data <- read.delim("../inputs/01.txt", sep = "", header = FALSE)
colnames(data) <- c("A", "B")
print(data)
```

```r
data.sorted <- data.frame(lapply(data, sort))
print(data.sorted)
```

```r
a <- data.sorted$A
b <- data.sorted$B
print(sum(abs(a - b)))
```
